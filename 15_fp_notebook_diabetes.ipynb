## Feature Engineering

Membuat fitur-fitur baru seperti kategori BMI, skor risiko kesehatan gabungan, dan fitur interaksi untuk tekanan darah tinggi dan kolesterol tinggi dari fitur yang sudah ada.

```python
import pandas as pd

# Re-load the data and apply previous cleaning steps to ensure df is defined.
# This is a defensive measure due to 'NameError: name 'df' is not defined'
# even though df was defined in previous steps.
df = pd.read_csv("diabetes_binary_health_indicators_BRFSS2015.csv")
df = df.drop_duplicates().copy()

# 1. Create 'BMI_Category' feature
bins = [0, 18.5, 24.9, 29.9, float('inf')]
labels = ['Underweight', 'Normal', 'Overweight', 'Obese']
df['BMI_Category'] = pd.cut(df['BMI'], bins=bins, labels=labels, right=True)

# 2. Create 'Health_Risk_Score'
# For PhysActivity, 0 means no physical activity (higher risk), 1 means physical activity (lower risk).
# To make it contribute positively to a risk score, we use (1 - PhysActivity).
risk_features = [
    'HighBP', 'HighChol', 'Smoker', 'Stroke', 'HeartDiseaseorAttack',
    'DiffWalk', 'GenHlth'
]
df['Health_Risk_Score'] = df[risk_features].sum(axis=1) + (1 - df['PhysActivity'])

# 3. Create 'HighBP_HighChol_Interaction'
df['HighBP_HighChol_Interaction'] = df['HighBP'] * df['HighChol']

print("New feature 'BMI_Category' created.")
print(df[['BMI', 'BMI_Category']].head())
print("\nNew feature 'Health_Risk_Score' created.")
print(df[['HighBP', 'HighChol', 'Smoker', 'Stroke', 'HeartDiseaseorAttack', 'PhysActivity', 'DiffWalk', 'GenHlth', 'Health_Risk_Score']].head())
print("\nNew feature 'HighBP_HighChol_Interaction' created.")
print(df[['HighBP', 'HighChol', 'HighBP_HighChol_Interaction']].head())
```

## Define and Train Classification Models

Define all classification models (Logistic Regression, Naive Bayes, Decision Tree, Voting Classifier, Random Forest, XGBoost). Create a reusable function to train each model on given training data, make predictions, and calculate performance metrics (accuracy, precision, recall, F1-score).


```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

def train_evaluate_model(model, X_train, y_train, X_test, y_test):
    """
    Trains a given model, makes predictions, and calculates performance metrics.
    """
    # Train the model
    model.fit(X_train, y_train)

    # Make predictions
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1] # Probability for the positive class

    # Calculate metrics using weighted average
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    roc_auc = roc_auc_score(y_test, y_proba)

    metrics = {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'roc_auc_score': roc_auc
    }
    return metrics

# Instantiate individual models
log_reg_model = LogisticRegression(solver='liblinear', random_state=42)
nai_bayes_model = GaussianNB()
decision_tree_model = DecisionTreeClassifier(random_state=42)
random_forest_model = RandomForestClassifier(random_state=42)
xgb_model = XGBClassifier(eval_metric='logloss', use_label_encoder=False, random_state=42)
adaboost_model = AdaBoostClassifier(random_state=42)

# Define the VotingClassifier to include all individual models
voting_clf = VotingClassifier(
    estimators=[
        ('lr', log_reg_model),
        ('gnb', nai_bayes_model),
        ('dt', decision_tree_model),
        ('rf', random_forest_model),
        ('xgb', xgb_model),
        ('ada', adaboost_model)
    ],
    voting='soft', # 'soft' voting uses predict_proba, 'hard' voting uses predict
    weights=[0.15, 0.1, 0.1, 0.2, 0.25, 0.2] # Adjust weights as needed
)

print("Classification models defined and `train_evaluate_model` function created.")
```

## Evaluate Models Across Scenarios

Iterate through each preprocessing scenario (engineered features, fixed feature selection, baseline) and for each scenario, train and evaluate all defined classification models. Collect all performance metrics in a structured format (e.g., a list of dictionaries).


```python
all_results = []

scenarios = {
    'Engineered Features': (X_train, X_test, y_train, y_test),
    'Fixed Feature Selection': (X_fixed_train, X_fixed_test, y_fixed_train, y_fixed_test),
    'Baseline': (X_baseline_train, X_baseline_test, y_baseline_train, y_baseline_test)
}

models = {
    'Logistic Regression': log_reg_model,
    'Naive Bayes': nai_bayes_model,
    'Decision Tree': decision_tree_model,
    'Random Forest': random_forest_model,
    'XGBoost': xgb_model,
    'AdaBoost': adaboost_model,
    'Voting Classifier': voting_clf
}

for scenario_name, (X_train_s_orig, X_test_s_orig, y_train_s_orig, y_test_s_orig) in scenarios.items():
    print(f"\nEvaluating models for scenario: {scenario_name}")

    # Create temporary copies to perform dropna and align y
    X_train_temp = X_train_s_orig.copy()
    y_train_temp = y_train_s_orig.copy()
    X_test_temp = X_test_s_orig.copy()
    y_test_temp = y_test_s_orig.copy()

    # Drop rows with NaNs from X_train_temp and align y_train_temp
    initial_train_rows = X_train_temp.shape[0]
    X_train_s = X_train_temp.dropna()
    y_train_s = y_train_temp.loc[X_train_s.index]
    rows_removed_train = initial_train_rows - X_train_s.shape[0]
    if rows_removed_train > 0:
        print(f"  Removed {rows_removed_train} NaN rows from training set for scenario '{scenario_name}'.")

    # Drop rows with NaNs from X_test_temp and align y_test_temp
    initial_test_rows = X_test_temp.shape[0]
    X_test_s = X_test_temp.dropna()
    y_test_s = y_test_temp.loc[X_test_s.index]
    rows_removed_test = initial_test_rows - X_test_s.shape[0]
    if rows_removed_test > 0:
        print(f"  Removed {rows_removed_test} NaN rows from testing set for scenario '{scenario_name}'.")


    for model_name, model_obj in models.items():
        print(f"  Training and evaluating {model_name}...")
        # For VotingClassifier, if using 'soft' voting, base estimators need predict_proba
        # Ensure base estimators are trained if they are not already. train_evaluate_model handles this.
        metrics = train_evaluate_model(model_obj, X_train_s, y_train_s, X_test_s, y_test_s)

        result = {
            'Scenario': scenario_name,
            'Model': model_name
        }
        result.update(metrics)
        all_results.append(result)

print("\nAll model evaluations complete.")
# Convert results to DataFrame for easier viewing and analysis
results_df = pd.DataFrame(all_results)
print(results_df)
```

Define scaling techniques (`MinMaxScaler`, `StandardScaler`, and 'No Scaling'), create a helper function to prepare data for each model's best configuration (scenario and sampling), evaluate these models with each scaling technique, and then compile and display the results in a comparison table.

## Define Scaling Techniques

Instantiate `MinMaxScaler` and `StandardScaler` from `sklearn.preprocessing` for normalization and standardization, respectively. Include 'No Scaling' as a baseline option.


```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaling_techniques = {
    'No Scaling': None,
    'MinMaxScaler': MinMaxScaler(),
    'StandardScaler': StandardScaler()
}

print("Scaling techniques dictionary defined and instantiated.")
```

## Prepare Data for Best Model Configurations with Sampling

Create a helper function that, for each model's best scenario and sampling technique, retrieves the appropriate `X_train`, `y_train`, `X_test`, `y_test` from the `scenarios` dictionary, applies duplicate removal and feature engineering/selection as needed, and then applies the specific best sampling technique to the training data. This ensures the data fed into the scaling step is consistent with the optimal configurations found previously.


```python
def prepare_data_for_model(model_row):
    """
    Retrieves and processes data (applying feature engineering/selection and sampling)
    based on the specified scenario and sampling technique in model_row.
    """
    scenario_name = model_row['Scenario']
    sampling_name = model_row['Sampling Technique']

    # Retrieve original data splits for the scenario
    X_train_s_orig, X_test_s_orig, y_train_s_orig, y_test_s_orig = scenarios[scenario_name]

    # Handle NaN values (defensive coding, although previous steps indicate no NaNs after cleaning)
    # Make copies to ensure original data in 'scenarios' remains untouched
    X_train_temp = X_train_s_orig.copy()
    y_train_temp = y_train_s_orig.copy()
    X_test_temp = X_test_s_orig.copy()
    y_test_temp = y_test_s_orig.copy()

    X_train_processed = X_train_temp.dropna()
    y_train_processed = y_train_temp.loc[X_train_processed.index]

    X_test_processed = X_test_temp.dropna()
    y_test_processed = y_test_temp.loc[X_test_processed.index]

    # Apply sampling technique if specified
    if sampling_name != 'No Sampling' and sampling_techniques.get(sampling_name) is not None:
        sampler = sampling_techniques[sampling_name]
        try:
            X_train_processed, y_train_processed = sampler.fit_resample(X_train_processed, y_train_processed)
            print(f"  Applied {sampling_name} to training data for scenario '{scenario_name}'.")
        except ValueError as e:
            print(f"    Skipping {sampling_name} due to error: {e}. Returning unsampled data.")

    return X_train_processed, X_test_processed, y_train_processed, y_test_processed

print("Helper function `prepare_data_for_model` defined.")
```

## Evaluate Models with Scaling Techniques

Iterate through each model listed in `best_f1_df` (which contains the best scenario and sampling technique for each model). For each model, obtain the pre-processed and sampled data using the helper function. Then, iterate through each defined scaling technique: apply the scaling to the (sampled) training and testing data, train the model, evaluate its performance using the `train_evaluate_model` function, and store all metrics in `scaling_results`.


```python
scaling_results = []

for index, row in best_f1_df.iterrows():
    model_name = row['Model']
    model_obj = models_for_sampling[model_name]

    print(f"\nProcessing model: {model_name} (Scenario: {row['Scenario']}, Sampling: {row['Sampling Technique']})")

    # Prepare data for the current model's best scenario and sampling
    X_train_processed, X_test_processed, y_train_processed, y_test_processed = prepare_data_for_model(row)

    for scaling_name, scaler in scaling_techniques.items():
        X_train_scaled = X_train_processed.copy()
        X_test_scaled = X_test_processed.copy()

        if scaler is not None:
            print(f"  Applying {scaling_name}...")
            # Fit scaler on training data and transform both train and test data
            scaler.fit(X_train_processed)
            X_train_scaled = pd.DataFrame(scaler.transform(X_train_processed), columns=X_train_processed.columns, index=X_train_processed.index)
            X_test_scaled = pd.DataFrame(scaler.transform(X_test_processed), columns=X_test_processed.columns, index=X_test_processed.index)

        print(f"    Training and evaluating {model_name} with {scaling_name}...")
        metrics = train_evaluate_model(model_obj, X_train_scaled, y_train_processed, X_test_scaled, y_test_processed)

        result = {
            'Model': model_name,
            'Scenario': row['Scenario'],
            'Sampling Technique': row['Sampling Technique'],
            'Scaling Technique': scaling_name
        }
        result.update(metrics)
        scaling_results.append(result)

print("\nAll scaling model evaluations complete.")
scaling_results_df = pd.DataFrame(scaling_results)
print(scaling_results_df.head())
```
